```python
import matplotlib.pyplot as plt
```

```python
   
if 'PBI_C' in data.columns and 'Año' in data.columns:
    promedio_pbi = data['PBI_C'].mean()
    data = data.sort_values(by='Año')
    plt.figure(figsize=(10, 6))
    plt.plot(data['Año'], data['PBI_C'], marker='o', color='blue', label='Crecimiento Económico')
    plt.axhline(y=promedio_pbi, color='red', linestyle='--', label=f'Promedio  = {promedio_pbi:.2f})')
    plt.title('Perú: Crecimiento económico, 1920 - 2021', fontsize=16)
    plt.xlabel('Año', fontsize=14)
    plt.ylabel('Crecimiento Económico', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=12)
    plt.show()
```

```python
if 'IPC promedio anual' in data.columns and 'Año' in data.columns:
    data['IPC promedio anual'] = pd.to_numeric(data['IPC promedio anual'], errors='coerce')
    data_filtrada = data[(data['Año'] >= 1995) & (data['Año'] <= 2021)]
    promedio_ipc = data_filtrada['IPC promedio anual'].mean()
    data_filtrada = data_filtrada.sort_values(by='Año')
    plt.figure(figsize=(10, 6))
    plt.plot(data_filtrada['Año'], data_filtrada['IPC promedio anual'], marker='o', color='red', label='Inflación')
    plt.axhline(y=promedio_ipc, color='blue', linestyle='--', label=f'Promedio = {promedio_ipc:.2f})')
    plt.title('Perú: Inflación, 1995-2021', fontsize=16)
    plt.xlabel('Año', fontsize=14)
    plt.ylabel('Inflación', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=12)
    plt.show()
```

```python
if 'Exportaciones de BBySS' in data.columns and 'Importaciones de BBySS' in data.columns:
    data['Exportaciones de BBySS'] = pd.to_numeric(data['Exportaciones de BBySS'], errors='coerce')
    data['Importaciones de BBySS'] = pd.to_numeric(data['Importaciones de BBySS'], errors='coerce')
    data['Exportaciones Netas'] = data['Exportaciones de BBySS'] - data['Importaciones de BBySS']
    tabla_comercio = data[['Año', 'Exportaciones de BBySS', 'Importaciones de BBySS', 'Exportaciones Netas']]
    tabla_comercio = tabla_comercio.sort_values('Año', ascending=False)
    tabla_comercio = tabla_comercio.round(2)
    print("Tabla de Comercio Exterior")
    print(tabla_comercio.to_string(index=False))
    tabla_comercio.to_csv('tabla_comercio_exterior.csv', index=False)
    print("\nLa tabla ha sido guardada en 'tabla_comercio_exterior.csv'")
else:
    print("Las columnas 'Exportaciones de BBySS' o 'Importaciones de BBySS' no existen en la base de datos.")
```

```python
if 'IPC promedio anual' in data.columns and 'IPC fin de periodo' in data.columns:
    data['IPC promedio anual'] = pd.to_numeric(data['IPC promedio anual'], errors='coerce')
    data['IPC fin de periodo'] = pd.to_numeric(data['IPC fin de periodo'], errors='coerce')
    data['Diferencia IPC'] = data['IPC fin de periodo'] - data['IPC promedio anual']
    tabla_ipc = data[['Año', 'IPC promedio anual', 'IPC fin de periodo', 'Diferencia IPC']]
    tabla_ipc = tabla_ipc.sort_values('Año', ascending=False)
    print(tabla_ipc.to_string(index=False))
    tabla_ipc.to_csv('tabla_ipc.csv', index=False)
    print("\nLa tabla ha sido guardada en 'tabla_ipc.csv'")
else:
    print("Las columnas 'IPC promedio anual' o 'IPC fin de periodo' no existen en la base de datos.")
```

```python
# 3. APIs

# Se realiza un análisis de la media movil a 12 meses.

BCRP_URL="https://estadisticas.bcrp.gob.pe/estadisticas/series/mensuales/resultados/PN01770AM/html"
BCRP=pd.read_html(BCRP_URL)
BCRP_data=BCRP[1]
BCRP_data
```

```python
errores = BCRP_data[BCRP_data['Fecha'].isna()]
print("Fechas no convertidas:\n", errores)
```

```python
#Cálculo del promedio móvil
BCRP_data['Fecha'] = pd.to_datetime(BCRP_data['Fecha'], format='%d%b', errors='coerce')
BCRP_data = BCRP_data.sort_values('Fecha')
BCRP_data['Promedio_Movil_12M'] = BCRP_data['Producto bruto interno y demanda interna (índice 2007=100) - PBI'].rolling(window=12).mean()
print(BCRP_data[['Fecha',  'Promedio_Movil_12M']].head(150))
```

```python
# Cálculo de la variación anual del promedio móvil de 12 meses
BCRP_data['Variacion_Anual_12M'] = BCRP_data['Promedio_Movil_12M'] - BCRP_data['Promedio_Movil_12M'].shift(12)
print(BCRP_data[['Fecha', 'Promedio_Movil_12M', 'Variacion_Anual_12M']].head(24))
```
